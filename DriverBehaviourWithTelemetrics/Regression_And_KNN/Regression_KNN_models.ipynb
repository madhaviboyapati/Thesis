## concatenate dataset

```python
dataset=pd.concat([data1,data2,data3,data4,data5])
dataset
dataset.to_csv("combined.csv")
```

```python
dataset['class']=dataset['Path'].map({"Lchange":1,"Rchange":2,"Lturn":3,"Rturn":4,"Straight":5})
dataset.tail()
```

```python
dataset.isnull().sum().sum()
```

## Splitting the dataset(train_test_split)

```python
X = dataset.loc[:,['acceleration_y','acceleration_x','acceleration_z','placement_x','placement_y','placement_z']]
y = dataset.loc[:,'class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

## MACHINE LEARNING MODELS

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 24, metric = 'minkowski', p = 2)
knn.fit(X_train, y_train)
```

```python
# Plotting a graph for n_neighbors 
from sklearn import metrics
from sklearn.metrics import accuracy_score
y_pred_knn = knn.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
```

```python
# Logistic Regression Algorithm
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state = 42)
logreg.fit(X_train, y_train)
```

### Making predictions on test dataset

```python
# Making predictions on test dataset
Y_pred_logreg = logreg.predict(X_test)
Y_pred_knn = knn.predict(X_test)
```

# MODEL EVALUATION

```python
# Evaluating using accuracy_score metric
from sklearn.metrics import accuracy_score
accuracy_logreg = accuracy_score(y_test, Y_pred_logreg)
accuracy_knn = accuracy_score(y_test, Y_pred_knn)
```

```python
# Accuracy on test set
print("Logistic Regression: " + str(accuracy_logreg * 100))
print("K Nearest neighbors: " + str(accuracy_knn * 100))
```
